```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv') #import Spotify data as dataframe
list(df) #get a list of headers
```

```python
#danceability vs. energy
df.plot(kind='scatter', x='loudness', y='energy')
plt.show()
```
